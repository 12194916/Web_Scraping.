Let's make it easy to get data everyday. So i wrote a function which does the job for a single link.

```python
def get_data(url):
  global request_time
  request_time = datetime.now()
  request_time = request_time.strftime("%H:%M:%S")
  result=requests.get(url)
  doc=BeautifulSoup(result.text,"html.parser")
  tags_p=doc.find_all("p")
  string_list=[]
  for i in tags_p:
    string_list.append(i.text)
    global string
    string = ' '.join((str(n) for n in string_list))
    
    for i in ['?','.','!','/',';',':','|','\xa0','	\n\n\n\n']:
      string = string.replace(i, "") 
      string = ''.join([i for i in string if not i.isdigit()])
      global word
      word=string.split(' ')
      word= list(dict.fromkeys(word))
      global counts
      counts=[]
      for j in word:
        count = string.count(j)
        counts.append(count)   
```

Then, we will see it for many links. And combine the dataframes at the end.

```python
def get_data_many(url_list):
  frames=[]
  for url in url_list:
    get_data(url)
    df=pd.DataFrame({"source_url":url,"access_datetime": request_time, "content":string,'word':word,'count':counts},index=range(len(word)))
    frames.append(df)
    global dataset
    dataset = pd.concat(frames)
  return dataset
```
